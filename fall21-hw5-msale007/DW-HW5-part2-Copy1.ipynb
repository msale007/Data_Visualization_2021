```python
df['Difference'] = df['POPESTIMATE2019']-df['POPESTIMATE2010']
df_new = df[['STNAME','CTYNAME', 'Difference']]
df_diff = df_new.groupby(['STNAME']).sum().reset_index()
df_diff.head()
```

```python
plt.figure(figsize=(20, 15))
plt.barh(df_diff['STNAME'], df_diff['Difference'] , color = 'rosybrown')
plt.title('Difference of population between 2010-2019',fontsize=20, color='black')
plt.xlabel('Difference', fontsize=20, color='black')
plt.ylabel('State', fontsize=20, color='black')
plt.grid()
plt.show()
```

```python
df_negative = df_diff[df_diff['Difference']<0]#.drop_duplicates()
df_negative
```

```python
plt.figure(figsize=(10, 5))
plt.barh(df_negative['STNAME'], df_negative['Difference'] , color = 'red')
plt.title('States With Negative Population Rate',fontsize=20, color='black')
plt.xlabel('Difference', fontsize=20, color='black')
plt.ylabel('State', fontsize=20, color='black')
plt.grid()
plt.show()
```
