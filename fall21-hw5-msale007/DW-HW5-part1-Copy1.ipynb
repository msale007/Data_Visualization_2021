```python
#ECDF function to generate x and y axis data
def ecdf(x):
    x_ecdf = np.sort(x)
    y_ecdf = np.arange(1, len(x) + 1) / len(x)
    return x_ecdf, y_ecdf

#Get the x and y data for ecdf plot from ecdf method
x,y = ecdf(df['CENSUS2010POP'])

#Plot the data using matplotlib
plt.plot(x, y, marker = '.', linestyle = 'none')
plt.title('eCDF for CENSUS2010POP ',fontsize=14, color='black')
plt.xlabel('CENSUS2010POP')
plt.ylabel('Percentage')
```

```python
#Import seaborn library and generate ecdf plot
import seaborn as sns
sns.ecdfplot(x = data['CENSUS2010POP'])
plt.title('eCDF for CENSUS2010POP using seaborn library ',fontsize=10, color='black')
```
